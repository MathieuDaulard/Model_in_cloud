```python
def import_data(path="DATA/archive/fruits-360_dataset/fruits-360/Training", max_categ=2):
    max_categ = max_categ
    dict_image = {}
    list_categ = []
    nb_dir = 0
    for i in os.scandir(path) :
        if nb_dir == max_categ:
            break
        if i.is_dir() or i.is_file():
            #print ('\n' + i.name + '\n')
            for j in os.scandir(os.path.join(path, i.name)):
                if j.is_dir() or j.is_file():
                    
                    list_categ.append((i.name, i.name + '/' + j.name))
                    
        nb_dir +=1
        
    return dict_image, list_categ

def import_datav2(path="DATA/archive/fruits-360_dataset/fruits-360/Training/", max_categ=2):
    max_categ = max_categ
    nb_dir = 0
    for i in os.scandir(path) :
        if nb_dir == max_categ:
            break
        elif (i.is_dir() or i.is_file()) and nb_dir == 0:
            df = spark.read.format("image").option("dropInvalid", True).load(os.path.join(path + i.name))
            df = df.withColumn("categ", lit(i.name))
        else:
            df1 = spark.read.format("image").option("dropInvalid", True).load(os.path.join(path + i.name))
            df1 = df1.withColumn("categ", lit(i.name))
            df = df.union(df1)
                    
        nb_dir +=1
        
    return df

def import_img(url):
    print(url)
    path="DATA/archive/fruits-360_dataset/fruits-360/Training/"
    size = 224, 224

    img = Image.open(os.path.join(path + url))
    img = img.resize(size) 
    
    # convert image object into array
    imageToMatrice = np.asarray(img).astype('float32')
    imageToMatrice = keras.applications.resnet_v2.preprocess_input(imageToMatrice)

    return imageToMatrice


def spark_image(url):
    path="DATA/archive/fruits-360_dataset/fruits-360/Training/"
    result = ImageSchema.toNDArray(os.path.join(path + url))
    return result
    
```

```python
def model_fn():
    model = ResNet50V2(wiegths = 'imagenet', include_top=False)
    return model

def preprocess(content):
    img = Image.open(content).resize([224,224])
    arr = img_to_array(img)
    return preprocess_input(arr)

def featurize_series(model, content_series):
    input = np.stack(content_series.map(preprocess))
    preds = model.predict(input)
    output = [p.flatten() for p in preds]
    return pd.Series(output)


def featurize_udf(content_series_iter):
    model = model_fn()
    for content_series in content_series_iter:
        yield featurize_series(model, content_series)
```
